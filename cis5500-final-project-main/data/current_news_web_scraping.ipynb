```python
!pip install newsapi-python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
from urllib.request import urlopen, Request
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from newsapi import NewsApiClient
from datetime import date, timedelta
```

```python
df = pd.read_csv('df_stock_info.txt')
```

```python
symbol_list = df['Ticker'].tolist()
```

```python
len(symbol_list)
```

```python
news_dict = {}
news_dict['Company'] = []
news_dict['Source'] = []
news_dict['Author'] = []
news_dict['Title'] = []
news_dict['URL'] = []
news_dict['Publish_Date'] = []
```

```python
%%time
for company in symbol_list:
  print('company', company)
  phrase = '{} stock'.format(company)
  newsapi = NewsApiClient(api_key='947a3edb5302483b8c10621da6cf461c')
  my_date = date.today() - timedelta(days = 31)
  articles = newsapi.get_everything(q=phrase,
                                  from_param = my_date.isoformat(),
                                  language="en",
                                  sort_by="relevancy",
                                  page_size = 5)

  for i in range(len(articles['articles'])):
    news_dict['Company'].append(company)
    news_dict['Source'].append(articles['articles'][i]['source']['name'])
    news_dict['Author'].append(articles['articles'][i]['author'])
    news_dict['Title'].append(articles['articles'][i]['title'])
    news_dict['URL'].append(articles['articles'][i]['url'])
    news_dict['Publish_Date'].append(articles['articles'][i]['publishedAt'])
```

```python
df_news = pd.DataFrame.from_dict(news_dict)
df_news
```
